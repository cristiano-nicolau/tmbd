## 4. About the main method/algorithm used [1,5 valor]

A **lógica fuzzy** foi adotada pela sua capacidade de lidar com incertezas típicas dos diagnósticos médicos, simulando o raciocínio humano e ajustando-se a variáveis contínuas, como glicose, pressão arterial e idade. Esta estratégia tem sido amplamente aplicada em diversos setores industriais, desde sistemas de controlo em edifícios inteligentes até eletrodomésticos de uso diário, demonstrando sua versatilidade na automação e otimização de processos.

Na avaliação do risco de diabetes, esta abordagem permite uma classificação numa escala contínua, sem limites rígidos, categorizando os dados em termos linguísticos, facilitando uma interpretação mais próxima da realizada por especialistas.

1. **Definição das Variáveis Fuzzy**:
    
    Cada variável de entrada—glicose, pressão arterial e idade—é categorizada em conjuntos fuzzy conforme descrito em baixo:
    
    - **Glicose (mg/dL)**: Indicador central no diagnóstico de diabetes, com os conjuntos fuzzy definidos como:
        - *Baixa*: 0-80 mg/dL
        - *Média*: 70–150 mg/dL
        - *Alta*: 140-200 mg/dL
        
    - **Pressão Arterial (mm Hg)**: A pressão diastólica é considerada devido à sua associação com riscos cardiovasculares, com os conjuntos fuzzy definidos como:
        - *Normal*: 0-90 mm Hg
        - *Elevada*: 75-115 mm Hg
        - *Alta*: 100-122 mm Hg
        
    - **Idade (anos)**: A idade é um importante fator de risco devido à diminuição da sensibilidade à insulina com o tempo. Os conjuntos fuzzy são:
        - *Jovem*: 0-40 anos
        - *Adulto*: 30-70 anos
        - *Idoso*: 60-100 anos


2. **Construção das Regras Fuzzy**

    O sistema fuzzy utiliza um conjunto de regras condicionais no formato "Se... então..." para gerar inferências. Estas regras integram múltiplas variáveis em simultâneo, resultando num diagnóstico mais contextualizado.
    

    #### Grupo 1: Glicose Baixa

    1. **Se** a glicose está *Baixa* e a pressão arterial está *Normal*, **então** o risco de diabetes é *Baixo*.
    2. **Se** a glicose está *Baixa* e a pressão arterial está *Elevada* ou *Alta* e a idade é *Jovem* ou *Adulto*, **então** o risco de diabetes é *Médio*.
    3. **Se** a glicose está *Baixa* e a pressão arterial está *Alta* e a idade é *Idoso*, **então** o risco de diabetes é *Alto*.

    #### Grupo 2: Glicose Média

    4. **Se** a glicose está *Média* e a pressão arterial está *Normal* e a idade é *Jovem* ou *Adulto*, **então** o risco de diabetes é *Médio*.
    5. **Se** a glicose está *Média* e a pressão arterial está *Normal* e a idade é *Idoso*, **então** o risco de diabetes é *Médio*.
    6. **Se** a glicose está *Média* e a pressão arterial está *Elevada*, **então** o risco de diabetes é *Alto*.
    7. **Se** a glicose está *Média* e a pressão arterial está *Alta* e a idade é *Adulto* ou *Idoso*, **então** o risco de diabetes é *Alto*.

    #### Grupo 3: Glicose Alta

    8. **Se** a glicose está *Alta* e a pressão arterial está *Normal*, **então** o risco de diabetes é *Alto*.
    9. **Se** a glicose está *Alta* e a pressão arterial está *Elevada* e a idade é *Adulto* ou *Idoso*, **então** o risco de diabetes é *Muito Alto*.
    10. **Se** a glicose está *Alta* e a pressão arterial está *Alta*, **então** o risco de diabetes é *Muito Alto*.


3. **Explicação das Regras Ajustadas**:

    As regras ajustadas classificam o risco de diabetes em Baixo, Médio, Alto e Muito Alto com base nas variáveis de entrada:

    - **Risco Baixo**: Todos os fatores estão dentro dos níveis considerados seguros, como glicose baixa, pressão arterial normal e idade jovem.
    - **Risco Médio**: Apenas um dos fatores apresenta valores moderados, como glicose média e pressão arterial elevada em indivíduos adultos.
    - **Risco Alto**: Dois fatores estão em níveis de risco, como glicose alta e pressão arterial elevada em pacientes  adultos.
    - **Risco Muito Alto**: Todos os fatores apresentam valores elevados, como glicose alta, pressão arterial elevada e idade avançada(idoso).

    Estas regras abrangem diferentes perfis clínicos e permitem uma avaliação detalhada e ajustada do risco de diabetes, promovendo uma interpretação precisa e contextualizada.

## 7. Main code as possible solution to the problem [1,5 valor] 

#### Fuzzificação das variáveis


```python
data_inicial = pd.read_csv('diabetes_dataset.csv')

# excluir as linhas nao usadas, e ficar apenas com as variáveis escolhidas
data = data_inicial[['Glucose', 'BloodPressure', 'Age', 'Outcome']]
#print(data.head())
#print(data.describe())
#print(data.info())

glucose_data = data['Glucose']
blood_pressure_data = data['BloodPressure']
age_data = data['Age']

glucose_range = np.linspace(0, 200, 100)
blood_pressure_range = np.linspace(0, 150, 100)
age_range = np.linspace(0, 100, 100)

# Fuzzyficacao

glucose_low = fuzz.trimf(glucose_range, [0, 0, 80])
glucose_medium = fuzz.trimf(glucose_range, [70, 110, 150])
glucose_high = fuzz.trimf(glucose_range, [140, 200, 200])

blood_pressure_normal = fuzz.trimf(blood_pressure_range, [0, 50, 90])
blood_pressure_elevated = fuzz.trimf(blood_pressure_range, [75, 95, 115])
blood_pressure_high = fuzz.trimf(blood_pressure_range, [100, 122, 122])

age_young = fuzz.trimf(age_range, [0, 0, 40])
age_middle_aged = fuzz.trimf(age_range, [30, 50, 70])
age_elderly = fuzz.trimf(age_range, [60, 85, 100])
```

View graphs of the fuzzy sets for each input variable.

```python
plt.figure()
plt.plot(glucose_range, glucose_low, label='Baixo')
plt.plot(glucose_range, glucose_medium, label='Medio')
plt.plot(glucose_range, glucose_high, label='Alto')
plt.title('Fuzzificação para Glicose')
plt.xlabel('Nível de Glicose')
plt.ylabel('Grau de Pertinência')
plt.legend()
plt.grid(True)

plt.figure()
plt.plot(blood_pressure_range, blood_pressure_normal, label='Normal')
plt.plot(blood_pressure_range, blood_pressure_elevated, label='Elevado')
plt.plot(blood_pressure_range, blood_pressure_high, label='Alto')
plt.title('Fuzzificação para Pressão Arterial')
plt.xlabel('Pressão Arterial (mm Hg)')
plt.ylabel('Grau de Pertinência')
plt.legend()
plt.grid(True)

plt.figure()
plt.plot(age_range, age_young, label='Jovem')
plt.plot(age_range, age_middle_aged, label='Adulto')
plt.plot(age_range, age_elderly, label='Idoso')
plt.title('Fuzzificação para Idade')
plt.xlabel('Idade (anos)')
plt.ylabel('Grau de Pertinência')
plt.legend()
plt.grid(True)

plt.show()
```

#### Regras Fuzzy

```python
glucose = ctrl.Antecedent(glucose_range, 'glucose')
blood_pressure = ctrl.Antecedent(blood_pressure_range, 'blood_pressure')
age = ctrl.Antecedent(age_range, 'age')
diabetes_risk = ctrl.Consequent(np.linspace(0, 100, 100), 'diabetes_risk')

# conjuntos fuzzy para a variável de saída
diabetes_risk['baixo'] = fuzz.trimf(diabetes_risk.universe, [0, 0, 25])
diabetes_risk['medio'] = fuzz.trimf(diabetes_risk.universe, [20, 50, 70])
diabetes_risk['alto'] = fuzz.trimf(diabetes_risk.universe, [60, 75, 100])
diabetes_risk['muito alto'] = fuzz.trimf(diabetes_risk.universe, [80, 100, 100])


glucose['baixo'] = glucose_low
glucose['medio'] = glucose_medium
glucose['alto'] = glucose_high

blood_pressure['normal'] = blood_pressure_normal
blood_pressure['elevado'] = blood_pressure_elevated
blood_pressure['alto'] = blood_pressure_high

age['jovem'] = age_young
age['adulto'] = age_middle_aged
age['idoso'] = age_elderly

# Grupo 1: Glicose Baixa
rule1 = ctrl.Rule(glucose['baixo'] & blood_pressure['normal'], diabetes_risk['baixo'])
rule2 = ctrl.Rule(glucose['baixo'] & (blood_pressure['elevado'] | blood_pressure['alto']) & (age['jovem'] | age['adulto']), diabetes_risk['medio'])
rule3 = ctrl.Rule(glucose['baixo'] & blood_pressure['alto'] & age['idoso'], diabetes_risk['alto'])

# Grupo 2: Glicose Média
rule4 = ctrl.Rule(glucose['medio'] & blood_pressure['normal'] & (age['jovem'] | age['adulto']), diabetes_risk['medio'])
rule5 = ctrl.Rule(glucose['medio'] & blood_pressure['normal'] & age['idoso'], diabetes_risk['medio'])
rule6 = ctrl.Rule(glucose['medio'] & blood_pressure['elevado'], diabetes_risk['alto'])
rule7 = ctrl.Rule(glucose['medio'] & blood_pressure['alto'] & (age['adulto'] | age['idoso']), diabetes_risk['alto'])

# Grupo 3: Glicose Alta
rule8 = ctrl.Rule(glucose['alto'] & blood_pressure['normal'], diabetes_risk['alto'])
rule9 = ctrl.Rule(glucose['alto'] & blood_pressure['elevado'] & (age['adulto'] | age['idoso']), diabetes_risk['muito alto'])
rule10 = ctrl.Rule(glucose['alto'] & blood_pressure['alto'], diabetes_risk['muito alto'])


# regras do sistema
rules = [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10] 

# Criar sistema de controle
diabetes_ctrl = ctrl.ControlSystem(rules)
diabetes_simulation = ctrl.ControlSystemSimulation(diabetes_ctrl)

```

#### Aplicação do algoritmo ao nosso dataset


```python
data = data.copy()

riscos = []
errors = 0

for index, row in data.iterrows():
    diabetes_simulation = ctrl.ControlSystemSimulation(diabetes_ctrl)

    # Replace missing blood pressure values with the mean value (this is a simple approach)
    if row['BloodPressure'] == 0:
        row['BloodPressure'] = int(blood_pressure_data.mean())
        

    diabetes_simulation.input['glucose'] = row['Glucose']
    diabetes_simulation.input['blood_pressure'] = row['BloodPressure']
    diabetes_simulation.input['age'] = row['Age']

    diabetes_simulation.compute()
    
    try:
        risco = diabetes_simulation.output['diabetes_risk']
        if risco is None:
            risco = 0
    except KeyError:
        #print(f"Erro ao calcular o risco na linha {index}")
        risco = 0
        errors += 1
    except Exception as e:
        print(f"Erro ao calcular o risco na linha {index}: {e}")
        risco = 0
        errors += 1

    riscos.append(risco)

if len(riscos) == len(data):
    data['DiabetesRisk'] = riscos
else:
    print("O número de riscos não corresponde ao número de entradas no DataFrame.")

data.to_csv('diabetes_dataset_fuzzy.csv', index=False)
```

## 10. Pros and cons of the approach [2,0 valor]

**Vantagens:**

1. **Capacidade para lidar com incertezas e ambiguidade:** A lógica fuzzy é especialmente eficaz para tratar problemas onde a incerteza e a ambiguidade estão presentes, uma vez que permite representar a realidade em valores intermediários entre o verdadeiro e o  falso.

2. **Flexibilidade e adaptabilidade:** A lógica fuzzy oferece flexibilidade no desenvolvimento de sistemas de controlo, permitindo adaptações conforme as necessidades de uma aplicação específica. Essa flexibilidade é particularmente útil em cenários onde é difícil definir fronteiras rígidas ou regras determinísticas, tornando-a uma abordagem versátil.

3. **Modelação intuitiva e uso de linguagem natural:** Uma das grandes vantagens da lógica fuzzy é a utilização de regras linguísticas, as quais facilitam a criação e interpretação do sistema. A possibilidade de empregar linguagem natural na construção de regras, simplifica o desenvolvimento e a manutenção do sistema, ao mesmo tempo que o torna acessível para especialistas de outras áreas.

**Desvantagens:**

1. **Complexidade na definição e ajuste de regras:** Para sistemas complexos, o número de regras necessárias para cobrir todas as situações possíveis pode crescer exponencialmente, tornando o sistema difícil de configurar e de ajustar. Esta complexidade aumenta a probabilidade de sobrecarga e de inconsistências nas regras, o que pode afetar negativamente o desempenho e a precisão do sistema.

2. **Dependência de conhecimento especializado:** O desenvolvimento de sistemas fuzzy eficazes depende frequentemente de conhecimento especializado para definir funções, ajustar regras, e calibrar parâmetros. Esta necessidade de especialização pode limitar a acessibilidade da abordagem, especialmente para utilizadores que não possuam experiência prévia em lógica fuzzy. Em casos como este trabalho, é necessário recorrer a documentação especializada para definir as regras e conjuntos fuzzy.

3. **Desafios de desempenho em sistemas de grande escala:** À medida que a complexidade do sistema aumenta, o tempo de processamento e a utilização de recursos computacionais também aumentam. A lógica fuzzy, particularmente em sistemas com múltiplas variáveis e regras complexas, pode requerer uma capacidade computacional significativa, o que se traduz em desafios de escalabilidade e desempenho em aplicações de grande dimensão.
